```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('Dataset.txt', sep ='\t')
```

```python
df.head()
```

```python
df.info()
```

```python
df = df.drop(columns = ['F15','F16'])
```

```python
X = df.drop(columns = 'C')
y = df['C']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
scaler = StandardScaler()
scaler.fit(X)
scaled_data = scaler.transform(X)
pca = PCA(n_components=5)
pca.fit(scaled_data)
x_pca = pca.transform(scaled_data)
```

```python
df_new = pd.DataFrame(data = x_pca, columns = ['PC1', 'PC2','PC3','PC4','PC5'])
df_new['C'] = y
```

```python
X = df_new.drop(columns = 'C')
y = df_new['C']
```

```python
df_new.head()
```

```python
df_new.describe()
```

```python
class_counts = y.value_counts()
print(class_counts)
```

```python
from imblearn.under_sampling import RandomUnderSampler

# Initialize the RandomUnderSampler with stratification
undersampler = RandomUnderSampler(sampling_strategy='auto', random_state=42)

# Apply undersampling
X_resampled, y_resampled = undersampler.fit_resample(X, y)

# Combine resampled features and target into a DataFrame
undersampled_data = pd.concat([pd.DataFrame(X_resampled, columns=X.columns), pd.Series(y_resampled, name='C')], axis=1)

# Print class distribution after undersampling
print("Class distribution after stratified undersampling:\n", undersampled_data['C'].value_counts())
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.svm import SVC
```

```python
clf = SVC()
X_train, X_test, y_train, y_test = train_test_split(undersampled_data.drop(columns = 'C'), undersampled_data['C'], test_size=0.2, random_state=42)
```

```python
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
```

```python
print(f'accuracy:{accuracy_score(y_test, y_pred)}')
print(f'precision:{precision_score(y_test, y_pred)}')
print(f'recall:{recall_score(y_test, y_pred)}')
print(f'f1_score:{f1_score(y_test, y_pred)}')
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
```

```python
rfc.fit(X_train, y_train)
y_pred2 = rfc.predict(X_test)
```

```python
print(f'accuracy:{accuracy_score(y_test, y_pred2)}')
print(f'precision:{precision_score(y_test, y_pred2)}')
print(f'recall:{recall_score(y_test, y_pred2)}')
print(f'f1_score:{f1_score(y_test, y_pred2)}')
```

```python
from xgboost import XGBClassifier
xgb = XGBClassifier(
    n_estimators=100,          # Number of trees
    learning_rate=0.1,         # Step size shrinkage
    max_depth=5,
)
```

```python
xgb.fit(X_train, y_train)
y_pred3 = xgb.predict(X_test)
```

```python
print(f'accuracy:{accuracy_score(y_test, y_pred3)}')
print(f'precision:{precision_score(y_test, y_pred3)}')
print(f'recall:{recall_score(y_test, y_pred3)}')
print(f'f1_score:{f1_score(y_test, y_pred3)}')
```
